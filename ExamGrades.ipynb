```python
import pandas as pd
import numpy as n
import matplotlib.pyplot as p
%matplotlib inline
p.style.use('mysty')
p.rcParams["figure.figsize"] = (12,12)
import figfun
import os
def col(i):
    return list(p.rcParams['axes.prop_cycle'])[i]['color']
```

# Wednesday 4-6 

```python
d = pd.read_csv('Wed46.csv')
ex = d['Exam 2']
print('Mean: {:.1f}\nStDv: {:.1f}\nMax : {:.0f}\nMin : {:.0f}'.format(ex.mean(),ex.std(),ex.max(),ex.min()))
```

```python
(ex//10*10).value_counts().sortlevel(ascending=False)
```

```python
ex.hist(bins=n.arange(50,110,10),rwidth=.8,color=col(0))
p.axis(ymax=15);
p.title('Wed 4-6');
```

#  Side by Side

```python
p.figure();
p.hist( [ex.values,ex2.values,ex3.values,ex4.values],
       bins=n.arange(50,110,10),range=(50,100),
       label=['W 4-6','W 6-8','Th 4-6','Th 6-8'] );
p.axis(ymax=15);
p.legend(loc='upper left');
p.title('All Sections');
```

```python
p.figure();
p.hist( [ex.values,ex2.values,ex3.values,ex4.values],
       bins=n.arange(50,110,10),range=(50,100),
       label=['W 4-6','W 6-8','Th 4-6','Th 6-8'],
      normed=True);
#p.axis(ymax=15);
p.legend(loc='upper left');
p.title('All Sections (normalized)');
```

# Whole class

```python
D = n.empty(0)
for i in [ex,ex2,ex3,ex4]:
    D = n.hstack((D,i.values))
print('Mean: {:.1f}\nStDv: {:.1f}\nMax : {:.0f}\nMin : {:.0f}'.format(n.nanmean(D),n.nanstd(D),n.nanmax(D),n.nanmin(D)))    
```

```python
print('Wed 4-6 Mean, Std:  {:.0f}, {:.0f}'.format(ex.mean(),ex.std()))
print('Wed 6-8 Mean, Std:  {:.0f}, {:.0f}'.format(ex2.mean(),ex2.std()))
print('Thu 4-6 Mean, Std:  {:.0f}, {:.0f}'.format(ex3.mean(),ex3.std()))
print('Thu 6-8 Mean, Std:  {:.0f}, {:.0f}'.format(ex4.mean(),ex4.std()))
```

```python
p.figure();
p.hist( D,
       bins=n.arange(20,110,10),range=(20,100),
       normed=False, rwidth=0.8,color=col(5));
p.title('Whole Class');
```
